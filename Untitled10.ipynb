# Find popular items globally


```python
df['Quantity'] = pd.to_numeric(df['Quantity'],errors='coerce')
grouped_data = df.groupby('Description')['Quantity'].sum().reset_index()
index_of_max = grouped_data['Quantity'].idxmax()
popular_items = grouped_data.loc[index_of_max]
popular_items
```

# Find popular items country-wise

```python
df['Quantity'] = pd.to_numeric(df['Quantity'],errors='coerce')
grouped_data = df.groupby(['Description','Country'])['Quantity'].sum().reset_index()
index_of_max = grouped_data['Quantity'].idxmax()
popular_item = grouped_data.loc[index_of_max]
popular_item
```

# Find Popular items month-wise

```python
df['Quantity'] = pd.to_numeric(df['Quantity'],errors='coerce')
grouped_data = df.groupby(['Description','InvoiceDate'])['Quantity'].sum().reset_index()
index_of_max = grouped_data['Quantity'].idxmax()
popular_items = grouped_data.loc[index_of_max]
popular_items
```

# Recommending Similar Products

```python
OnlineRetail = df.pivot_table(index=['InvoiceNo'],columns=['Description'],values='Quantity')
OnlineRetail.head()
```

```python
Quantity.sort_values('No_of_purchases',ascending=False).head(10)
```

```python
Quantity.head()
```

```python
PartyBunting = OnlineRetail['PARTY BUNTING']
LunchBag = OnlineRetail['LUNCH BAG RED RETROSPOT']
PartyBunting.head()
```

```python
similar_to_PartyBunting=OnlineRetail.corrwith(PartyBunting)
similar_to_LunchBag = OnlineRetail.corrwith(LunchBag)
similar_to_PartyBunting.head(10)
```

```python
corr_PartyBunting = pd.DataFrame(similar_to_PartyBunting,columns=['Correlation'])
corr_PartyBunting.dropna(inplace=True)
corr_PartyBunting = corr_PartyBunting.join(Quantity['No_of_purchases'])
corr_PartyBunting.head()
```

```python
corr_PartyBunting[corr_PartyBunting['No_of_purchases']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_LunchBag = pd.DataFrame(similar_to_LunchBag,columns=['Correlation'])
corr_LunchBag.dropna(inplace=True)
corr_LunchBag = corr_LunchBag.join(Quantity['No_of_purchases'])
corr_LunchBag[corr_LunchBag['No_of_purchases']>100].sort_values('Correlation',ascending=False).head()
```
